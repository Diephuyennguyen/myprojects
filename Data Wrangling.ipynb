# Problem: Calculate revenue for each product. Noted that price products changed usually. 

```python
import pandas as pd
import datetime as dt
sales=pd.read_csv('sales.csv')
prices=pd.read_csv('prices.csv')
print(sales.head(10))
print(prices.head(10))
```

```python
sales.ordered_at=pd.to_datetime(sales['ordered_at'])
prices.updated_at=pd.to_datetime(prices['updated_at'])
sales=sales.sort_values('ordered_at')
prices=prices.sort_values('updated_at')
```

```python
merge=pd.merge_asof(sales,prices,left_on='ordered_at',right_on='updated_at',by='product_id',direction='nearest')
list_price=[]
for i in merge.index: 
    if merge.ordered_at[i]<merge.updated_at[i]:
        list_price.append(merge.old_price[i])
    else:
        list_price.append(merge.new_price[i])
merge['price']=list_price
merge['sales']=merge['quantity_ordered']*merge['price']
merge    
```

```python
merge.groupby('product_id')['sales'].sum()
```
